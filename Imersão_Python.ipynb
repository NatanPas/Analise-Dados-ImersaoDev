Aula 2 - Importação

```python
import pandas as pd
import plotly.express as px
```

```python
!wget https://github.com/NatanPas/Analise-Dados-ImersaoDev/raw/main/Tabela%20de%20a%C3%A7%C3%B5es.xlsx
```

```python
df_principal = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
df_Total_de_acoes = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticker = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Ticker")
df_Ticker
```

```python
df_Chat = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name="Chat")
df_Chat
```

```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_Total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Status'] = pd.Series(['Subiu' if x > 0 else 'Desceu' if x < 0 else 'Estavel' for x in df_principal['variacao_rs']])
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_Chat, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
df_principal['Cat_idade'] = pd.Series(['Mais de 100 anos' if x > 100 else 'Menos que 50 anos' if x < 50 else 'Entre 50 a 100 anos' for x in df_principal['Idade']])
df_principal
```

```python
# 1. Calcular o maior dia, o menor dia e a média do dia
maior_dia = df_principal['variacao_rs'].max()
menor_dia = df_principal['variacao_rs'].min()
media_dia = df_principal['variacao_rs'].mean()

# 2. Calcular a média de quem subiu e a média de quem desceu
media_subiu = df_principal[df_principal['Status'] == 'Subiu']['variacao_rs'].mean()
media_desceu = df_principal[df_principal['Status'] == 'Desceu']['variacao_rs'].mean()

# Imprimir os resultados
print("Maior dia:\tR$", maior_dia)
print("Menor dia:\tR$", menor_dia)
print("Média dia:\tR$", media_dia)
print("Média de quem subiu:\tR$", media_subiu)
print("Média de quem desceu:\tR$", media_desceu)
```

```python
df_principal_subiu = df_principal[df_principal['Status'] == 'Subiu']
df_principal_subiu
```

```python
df_principal_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_principal_segmento
```

```python
df_principal_saldo = df_principal.groupby('Status')['variacao_rs'].sum().reset_index()
df_principal_saldo
```

```python
fig = px.bar(df_principal_saldo, x='Status', y='variacao_rs', text='variacao_rs', title='Variação reais por Status')
fig.show()
```
